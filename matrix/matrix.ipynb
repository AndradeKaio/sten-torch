```python
import matplotlib.pyplot as plt
from matrix_bench import parse_from_file
```

```python
sizes = [3072, 4096, 5096, 6096, 7096, 8096, 9096, 10096, 11096]
```

```python
cpu_dense_nocomp, gpu_dense_nocomp = parse_from_file("nocomp-dense.txt")
cpu_dense_comp, gpu_dense_comp = parse_from_file("comp-dense.txt")
cpu_sparse_nocomp, gpu_sparse_nocomp = parse_from_file("nocomp-sparse.txt")
cpu_sparse_comp, gpu_sparse_comp = parse_from_file("comp-sparse.txt")
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 10))
axs[0].plot(sizes, cpu_dense_nocomp, color="red", label="dense-nocomp")
axs[0].plot(sizes, cpu_dense_comp, color="green", label="dense-comp")
axs[0].plot(sizes, cpu_sparse_nocomp, color="red", linestyle="--", label="sparse-nocomp")
axs[0].plot(sizes, cpu_sparse_comp, color="green", linestyle="--", label="sparse-comp")
axs[0].set_title("CPU")
axs[0].set_xlabel("Matrix Size (Squared)")
axs[0].set_ylabel("Time (s)")
axs[0].grid(True)
axs[0].legend()

axs[1].plot(sizes, gpu_dense_comp, color="green", label="dense-comp")
axs[1].plot(sizes, gpu_dense_nocomp, color="red", label="dense-nocomp")
axs[1].plot(sizes, gpu_sparse_nocomp, color="red", linestyle="--", label="sparse-nocomp")
axs[1].plot(sizes, gpu_sparse_comp, color="green", linestyle="--", label="sparse-comp")
axs[1].set_title("GPU")
axs[1].set_xlabel("Matrix Size (Squared)")
axs[1].set_ylabel("Time (s)")
axs[1].grid(True)
axs[1].legend()
plt.tight_layout()
plt.savefig('big-matrices.png')
```
